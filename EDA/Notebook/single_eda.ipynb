Single value based EDA

```python
df_singlevalues = df[['session', 'participant', 'gesture', 'iemg', 'mav', 'ssi', 'myopulse', 'wflen',
       'diffvar', 'dasd', 'willison']]
df_singlevalues
```

```python
grouped = df_singlevalues.groupby(df_singlevalues.session)
df_svsession1 = grouped.get_group(1)
df_svsession2 = grouped.get_group(2)
df_svsession3 = grouped.get_group(3)
```

```python
avg_session = df_singlevalues.groupby('gesture').mean()
avg_session = avg_session[['iemg', 'mav', 'ssi', 'myopulse', 'wflen', 'diffvar', 'dasd', 'willison']]
avg_session
```

```python
avg_session1 = df_svsession1.groupby('gesture').mean()
avg_session1 = avg_session1[['iemg', 'mav', 'ssi', 'myopulse', 'wflen', 'diffvar', 'dasd', 'willison']]
avg_session1
```

```python
avg_session2 = df_svsession2.groupby('gesture').mean()
avg_session2 = avg_session2[['iemg', 'mav', 'ssi', 'myopulse', 'wflen', 'diffvar', 'dasd', 'willison']]
avg_session2
```

```python
avg_session3 = df_svsession3.groupby('gesture').mean()
avg_session3 = avg_session3[['iemg', 'mav', 'ssi', 'myopulse', 'wflen', 'diffvar', 'dasd', 'willison']]
avg_session3
```
